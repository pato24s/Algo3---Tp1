```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
bt=pd.read_csv('./ejer_2salida15a45.csv')
dp=pd.read_csv('./ejer_3salida15a45.csv')
btMean=bt.mean()
dpMean=dp.mean()
df = pd.DataFrame({'Backtracking c/Poda':btMean,'Programacion Dinamica' :dpMean})
df.plot(logy=True)
plt.title('Tiempo en funcion de la longitud de la entrada')
plt.ylabel('Tiempo (s)')
plt.xlabel('Longitud de la entrada')
plt.show()

```
