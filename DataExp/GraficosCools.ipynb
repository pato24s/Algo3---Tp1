```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
ran=pd.read_csv('./ejer_1salida1a15.csv')
crc=pd.read_csv('./ejer_1salida1a15Creciente.csv')
dec=pd.read_csv('./ejer_1salida1a15Decreciente.csv')
df = pd.DataFrame({'Random':ran.mean(),'Creciente':crc.mean(),'Decreciente':dec.mean()})
df.plot()
plt.title('Tiempo en funcion de la longitud de la entrada - Backtracking')
plt.ylabel('Tiempo (s)')
plt.xlabel('Longitud de la entrada')
plt.show()
```

```python
ran=pd.read_csv('./ejer_2salida1a15.csv')
crc=pd.read_csv('./ejer_2salida1a15Creciente.csv')
dec=pd.read_csv('./ejer_2salida1a15Decreciente.csv')
dfpd = pd.DataFrame({'Random':ran.mean(),'Creciente':crc.mean(),'Decreciente':dec.mean()})
dfpd.plot()
plt.title('Tiempo en funcion de la longitud de la entrada - Backtracking c/Poda')
plt.ylabel('Tiempo (s)')
plt.xlabel('Longitud de la entrada')
plt.show()
```
